## Rating every variable of the dataset as 0 and 1

```python
## country column

def country(col):
    if col == "Uganda":
        return 0
    elif col == "Rwanda":
        return 1
    elif col == "Tanzania":
        return 1
    elif col == "Kenya":
        return 1
    
train_data['country'] = train_data['country'].apply(country)
test_data['country'] = test_data['country'].apply(country)
```

```python
train_data
```

```python
## year column

def year(col):
    if col == 2016:
        return 1
    elif col == 2017:
        return 2
    elif col == 2018:
        return 3
    
train_data['year'] = train_data['year'].apply(year)
test_data['year'] = test_data['year'].apply(year)
train_data
```

```python
## bank account column

def acct(col):
    if col == "No":
        return 0
    elif col == "Yes":
        return 1
    
train_data['bank_account'] = train_data['bank_account'].apply(acct)
train_data
```

```python
## location_type column

def location(col):
    if col == "Rural":
        return 0
    elif col == "Urban":
        return 1
    
train_data['location_type'] = train_data['location_type'].apply(location)
test_data['location_type'] = test_data['location_type'].apply(location)
train_data
```

```python
## cellphone access column, using the acct function

train_data['cellphone_access'] = train_data['cellphone_access'].apply(acct)
test_data['cellphone_access'] = test_data['cellphone_access'].apply(acct)
train_data
```

```python
## respondent gender column

def gender(col):
    if col == "Male":
        return 0
    elif col == "Female":
        return 1
    
train_data['gender_of_respondent'] = train_data['gender_of_respondent'].apply(gender)
test_data['gender_of_respondent'] = test_data['gender_of_respondent'].apply(gender)
train_data
```

```python
## relationship_with_head column

def household(col):
    if col == "Other non-relatives":
        return 0
    elif col == "Other relative":
        return 1
    elif col == "Parent":
        return 2
    elif col == "Child":
        return 3
    elif col == "Spouse":
        return 4
    elif col == "Head of Household":
        return 5
    
train_data['relationship_with_head'] = train_data['relationship_with_head'].apply(household)
test_data['relationship_with_head'] = test_data['relationship_with_head'].apply(household)
train_data
```

```python
## marital status column

def marital(col):
    if col == "Dont know":
        return 0
    elif col == "Divorced/Seperated":
        return 1
    elif col == "Widowed":
        return 2
    elif col == "Single/Never Married":
        return 3
    elif col == "Married/Living together":
        return 4
    elif col == "Head of Household":
        return 5
    
train_data['marital_status'] = train_data['marital_status'].apply(marital)
test_data['marital_status'] = test_data['marital_status'].apply(marital)
train_data
```

```python
sum(train_data.isnull().values.ravel())
```

```python
## education level column

def education(col):
    if col == "Other/Dont know/RTA":
        return 0
    elif col == "No formal education":
        return 1
    elif col == "Vocational/Specialised training":
        return 2
    elif col == "Primary education":
        return 3
    elif col == "Secondary education":
        return 4
    elif col == "Tertiary education":
        return 5
    
train_data['education_level'] = train_data['education_level'].apply(education)
test_data['education_level'] = test_data['education_level'].apply(education)
train_data
```

```python
## job_type column

def job_type(col):
    if col == "Dont Know/Refuse to answer":
        return 1
    elif col == "No Income":
        return 2
    elif col == "Government Dependent":
        return 3
    elif col == "Other Income":
        return 4
    elif col == "Formally employed Government":
        return 5
    elif col == "Formally employed Private":
        return 6
    elif col == "Remittance Dependent":
        return 7
    elif col == "Farming and Fishing":
        return 8
    elif col == "Informally employed":
        return 9
    elif col == "Self employed":
        return 10
    
train_data['job_type'] = train_data['job_type'].apply(job_type)
test_data['job_type'] = test_data['job_type'].apply(job_type)
train_data
```
